```python
def get_subjects_with_stories(participants_df, stories):
    """
    Return list of subject IDs that have all stories in the list.
    
    Inputs: 
    participants_df: Pandas DataFrame with Narratives subject metadata
    stories: List with names of stories of interest
    
    Outputs:
    participants_list: List with names of the subjects that listened to these stories
    """
    
    def has_stories(row, stories):
        if pd.isna(row['task']):
            return False
        tasks = row['task'].split(',')
        return all(s in tasks for s in stories)
    
    mask = participants_df.apply(has_stories, axis=1, stories=stories) # apply has_stories to each row
    selected = participants_df[mask] # select rows where the mask is true
    
    participants_list = selected['participant_id'].tolist()
    
    return participants_list # return the subject ids as a list
```

```python
stories = ['piemanpni', 'bronx']
subjects = get_subjects_with_stories(participants, stories)
print(subjects)
```
